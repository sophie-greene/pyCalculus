```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Fx(x,r,A):
    return r*(x)*(1-(x/A))
```

```python
def gen(t0,r,A):
    y=np.zeros(len(t))
    y[0]=t0
    for i in range(1,len(t)):
        y[i]=Fx(y[i-1],r,A)
    return y
```

```python
def lyapunov(x,d0,lmb):
    y=np.asarray([d0*pow(2,i*lmb) for i in x])
    return y
```

```python
t0=.1
r=4.0
A=1
it=5
init=.2
t = range(0,it)
d0=1e-2
y=gen(init+d0,r,A)
z=gen(init,r,A)
w=lyapunov(t,d0,1)
plt.figure(figsize=(20,5))
plt.plot(t,y,'b',label='Iniatial +e-16')
plt.plot(t,z,'r',label='Iniatial')
plt.legend((['Iniatial','Iniatial +e-16']))
plt.show()
plt.figure(figsize=(20,5))
plt.plot(t,abs(y-z),'g')
plt.plot(t,w,'b')
plt.legend((['abs(Iniatial - Iniatial +e-16)','lyabunov']))
plt.show()
#print y[-1],z[-1]
```

```python
fig=plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
l=plt.plot(range(-20,120),Fx(np.asarray(range(-20,120),dtype=np.float),2.0,100.0))
l3=plt.plot(range(-20,120),Fx(np.asarray(range(-20,120),dtype=np.float),3.0,100.0))
l4=plt.plot(range(-20,120),Fx(np.asarray(range(-20,120),dtype=np.float),3.0,50.0))
plt.axis([-20,110,-20,80])
plt.suptitle('Logistic Differential Equation', fontsize = 18)
ax.set_xlabel('time t', fontsize = 16)
ax.set_ylabel('Growth rate dP/dt', fontsize = 16)
#ax.axhline(y=0, color='k')
#ax.axvline(x=0, color='k')


# set the x-spine (see below for more info on `set_position`)
ax.spines['left'].set_position('zero')

# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

# set the y-spine
ax.spines['bottom'].set_position('zero')
ax.margins(0,0)
# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()
plt.setp(l, 'linewidth',2)
plt.setp(l, 'linestyle','--') 
plt.setp(l3, 'linewidth',2)
plt.setp(l3, 'linestyle','-') 
plt.setp(l4, 'linewidth',2)
plt.setp(l4, 'linestyle','-.') 
plt.legend(['r=2 k=100','r=3 k=100','k=3 k=50'])
plt.grid()
```
