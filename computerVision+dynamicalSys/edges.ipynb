```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
img1 = cv2.imread('frizzy.png',0)
edges1 = cv2.Canny(img1,100,200)
img2 = cv2.imread('froom.png',0)
edges2 = cv2.Canny(img2,100,200)
plt.figure(figsize=(20,20))
plt.subplot(151),plt.imshow(img1,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(152),plt.imshow(edges1,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.subplot(153),plt.imshow(img2,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(154),plt.imshow(edges2,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(155),plt.imshow(edges1&edges2,cmap = 'gray')
plt.title('common Edges'), plt.xticks([]), plt.yticks([])
plt.tight_layout();

plt.show()
```
