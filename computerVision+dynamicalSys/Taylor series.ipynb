```python
def tanx(n):
    return [tan(x) for x in n]
def sinx(n):
    return [1*sin(x) for x in n]
def cosx(n):
    return [1*cos(x) for x in n]
t0=-5*np.pi
tn=5*np.pi
dt=.01
# Define discretized time ; assumes dt divides nicely into T
tt = np.linspace (t0 , tn, int(tn/dt )+1)
yt=tanx(tt)
ys=sinx(tt)
yc=cosx(tt)
ax  = plt.figure(figsize=(20,8)).add_subplot(111)
l=plt.plot(tt,yt);
l1=plt.plot(tt,ys);
l2=plt.plot(tt,yc);
plt.setp(l,'linestyle','--')
plt.setp(l1,'linewidth',3)
plt.setp(l2,'linewidth',3)
t_tick,t_label =create_pi_labels(t0/np.pi, tn/np.pi,.5 )
ax.set_xticks(t_tick)
ax.set_yticks(np.linspace (-3 , 3, int(3/.25 )+1))
ax.set_xticklabels(t_label, fontsize=18)
plt.ylim((-3,3))
plt.xlim((t0,tn))
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.yticks(fontsize=16)
ax.legend(['$tan\ x$','$sin\ x$ ','$cos\ x$ '],fontsize=20)
plt.ylabel('Value',fontsize=18)
plt.xlabel('Time (radian)',fontsize=18)
tit=plt.title(r"Trigonometric Relationship: $(tan\ x=\frac{sin\ x}{cos\ x})$",fontsize=24)
tit.set_y(1.09) 
plt.grid()
plt.show()
ax  = plt.figure(figsize=(20,8)).add_subplot(111)
l=plt.plot(tt,yt);
l1=plt.plot(tt,ys);
l2=plt.plot(tt,yc);
plt.setp(l,'linestyle','--')
plt.setp(l1,'linewidth',2)
plt.setp(l2,'linewidth',2)

t_tick,t_label =create_pi_labels(t0/np.pi, tn/np.pi,.5 )
ax.set_xticks(t_tick)
ax.set_yticks(np.linspace (100 ,-100, int(100/10.0 )+1))
ax.set_xticklabels(t_label, fontsize=18)

plt.xlim((t0,tn))
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.yticks(fontsize=16)
ax.legend(['$tan\ x$','$sin\ x$','$cos\ x$'],fontsize=20)
plt.ylabel('Value',fontsize=18)
plt.xlabel('Time (radian)',fontsize=18)

plt.grid()
```

```python
def fx(x):
    return np.sin(x)
def fxa(d):
    x=Symbol('x')
    y=series(sin(x),x,0,100)
    l=str(y)
    m=l.split('+')
    y=sympify('+'.join(m[:-1]))
    return [y.evalf(subs={x:i}) for i in d]
```

```python
t0=-30
tn=30
dt=.05
# Define discretized time ; assumes dt divides nicely into T
t = np.linspace (t0 , tn, int(tn/dt )+1)
```

```python
yx=fx(t)
yxa=fxa(t)
```

```python

ax  = plt.figure(figsize=(20,8)).add_subplot(111)

ax.plot(t,yx,'r')
l=ax.plot(t,yxa,'g')

t_tick,t_label =create_pi_labels(-4, 4,.5 )
ax.set_xticks(t_tick*np.pi)
ax.set_xticklabels(t_label, fontsize=20)
plt.axhline(y=0)
plt.axvline(x=0)
plt.setp(l, 'linewidth',2)
plt.setp(l, 'linestyle','--') 
ax.legend(['$sin (x)$','$\sum_{n=0}^{80} a_n x^n$ '],fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid()

plt.show()
```
