```python
def fe(n):
    x=Symbol('x')
    y=sympify('(x**(2/3))*(exp(x))')
    
    return [(y.evalf(subs={x:i})) for i in n]
```

```python
plt.figure(figsize=(15,5))
t= np.linspace (-5 ,.1, int(100/.25)+1)
y=fe(t)
res=[re(k) for k in y]
ims=[im(k) for k in y]
mg=[sqrt(re(k)**2+im(k)**2)for k in y]
plt.plot(t,res)
plt.plot(t,ims)
plt.plot(t,mg)
plt.legend(['real','imag','magnitude'])
```

```python
def fx3(n):
    x=Symbol('x')
    y=sympify('x**3-3*x')
    
    return [(y.evalf(subs={x:i})) for i in n]
def fxp3(n):
    x=Symbol('x')
    y=sympify('3*x**2-3')
    
    return [(i/abs(i))*(y.evalf(subs={x:i})) if i!=0 else 0 for i in n]
```

```python
plt.figure(figsize=(15,5))
t= np.linspace (-3 ,3, int(2/.01)+1)
y=fx3(t)
ddy=fxp3(t)
res=[re(k) for k in y]
ims=[im(k) for k in y]
mg=[sqrt(re(k)**2+im(k)**2)for k in y]
plt.plot(t,y)
plt.plot(t,ddy)
plt.plot([0],fxp3([0]),'x')
plt.legend(['fx','fpx','magnitude'])
plt.ylim([-3.5,3.5])
plt.grid()
```
