```python
a=Symbol('a')
C1=Symbol('C1')
t=Symbol('t')
T = Function('T')
lf = diff(T(t), t) - a*(T(t)-20)
```

#### solve for alfa using the original dT/dt at t=0 is -5

```python
fp=Lambda(t,a*(T(t)-20))
alfa=solve(fp(0).subs(T(0),70)+5,a)[0]
lf=lf.subs(a,alfa)
```

```python
s=dsolve(lf,T(t))
```

###### substitute t=0 in the T(t) resulting from solving the ode

```python
Eq(s.subs(t,0),70)
```

```python
f = Lambda((t),s.rhs )
```

find the value of C1 by solving the equation T(0)-70=0 for the value of C1, this is from the knowldge that T(0)=70
    then substitute in T

```python
c=solve(f(0)-70,C1)[0]
```

```python
s=s.subs(C1,c)
```

```python
fT=Lambda((t),s.rhs )
```

```python
## plot the solution
ttt=np.linspace(0,100,int(100/.01)+1)
yyy=[fT(i).evalf() for i in ttt]
plt.plot(ttt,yyy);
plt.grid();
plt.ylim([0,70]);
```

```python
plt.figure(figsize=(15,5))
t=Symbol('t')
C1=Symbol('C1')
m=Symbol('m')
y = Function('y')
fff = diff(y(t), t)-y(t)+t
sol=dsolve(fff,y(t))

rngr=[[-1,0],[0,0],[1,0],[0,1]]
strr=[]
f=Lambda (t,sol.rhs)
c1=solve(f(0)-0,C1)[0]
   
s=sol.subs(C1,c1)
for tt,yy in rngr:
    #7print tt,yy
    
    f=Lambda (t,s.rhs)
    fp=Lambda ((m,t),m-t)
    rng=xrange(-1,1)

    yr=np.array([[fp(f(i).evalf(),i),i] for i in rng])
    #print c1
    
    l=plt.plot(yr[0,:],yr[1,:])
    plt.setp(l,'linewidth',3)
    strr.append("$"+latex(s)+"$");
plt.legend(strr);
plt.ylim([-3,3])
plt.xlim([-3,5])
plt.grid();
plt.axhline(y=0,color='k');
plt.axvline(x=0,color='k');
```
